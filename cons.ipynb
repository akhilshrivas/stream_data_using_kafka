```python
# consumer.py
from kafka import KafkaConsumer

consumer = KafkaConsumer(
    'test',
    bootstrap_servers='localhost:9092',
    auto_offset_reset='earliest',
    enable_auto_commit=True,
    group_id='csv-consumer'
)

print("Listening for messages... Press Ctrl+C to stop.")

try:
    for message in consumer:
        print(f"Received: {message.value.decode('utf-8')}")
except KeyboardInterrupt:
    print("\nStopping consumer politely...")

finally:
    consumer.close()
    print("Consumer closed.")
```

```python
# streamlit_consumer.py
import streamlit as st
from kafka import KafkaConsumer

consumer = KafkaConsumer(
    'test',
    bootstrap_servers='localhost:9092',
    auto_offset_reset='earliest',
    enable_auto_commit=True,
    group_id='streamlit-dashboard'
)

st.title("Real-time Kafka Dashboard")

placeholder = st.empty()

data = []

for message in consumer:
    value = message.value.decode('utf-8')
    data.append(value)

    with placeholder.container():
        st.write("Latest Data:", value)
        st.line_chart(data)  # simple line chart
```

```python
import streamlit as st
from kafka import KafkaConsumer
import pandas as pd

# Kafka consumer
consumer = KafkaConsumer(
    'salary-topic',
    bootstrap_servers='localhost:9092',
    auto_offset_reset='earliest',
    enable_auto_commit=True,
    group_id='salary-dashboard'
)

st.title("IT Salary Survey Dashboard")

placeholder = st.empty()
data = []

for message in consumer:
    row = message.value.decode('utf-8').split(',')
    data.append(row)

    # Convert to DataFrame with some selected columns (adjust based on your CSV)
    df = pd.DataFrame(data, columns=[
        "Timestamp", "Employer", "Location", "Position", "YearsExperience", "Salary"
    ][:len(row)])  # prevent column mismatch

    with placeholder.container():
        st.subheader("Live Data Table")
        st.dataframe(df.tail(10))  # show last 10 rows

        if "Salary" in df.columns:
            st.subheader("Salary Distribution")
            df["Salary"] = pd.to_numeric(df["Salary"], errors="coerce")
            st.bar_chart(df["Salary"].dropna())
```
